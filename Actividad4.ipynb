```python
import random

def main():
    while True:
        try:
            print("Tareas:")
            print("1.- Generar 40 numeros aleatorios (0-200) y mostrar si son par o impar")
            print("2.- Mostrar tabla de un numero")
            print("3.- Dar calificacion ahsta que sea valida")
            print("4.- Introducir numeros que no sean 0")
            print("5.- 3 calificaciones")
            print("6.- dar numeros y mostrar media, minimo y mayor")
            print("7.- Genera 15 numeros impares o maximo de 25 numeros, mostrar media de pares e impares")
            print("8.- Validar numero en rango dado, mostrar numeros y promedio")
            print("9.- Recibir valor de triangulo y calcular area (funcion)")
            print("10.- Validar un numero en un rango dado")
            print("11.- Salir del programa")
            op = int(input("Su seleccion: "))
        except ValueError:
            print("Elija un numero\n")
        else:
            match op:
                case 1:
                    act1()
                    print("")

                case 2:
                    act2()
                    print("")

                case 3:
                    act3()
                    print("")

                case 4:
                    act4()
                    print("")

                case 5:
                    act5()
                    print("")

                case 6:
                    act6()
                    print("")

                case 7:
                    act7()
                    print("")

                case 8:
                    while True:
                        try:
                            LimIn,LimMax = 0,0
                            LimIn = int(input("Dame el limite inferior: "))
                            LimMax = int(input("Dame el limite maximo: "))
                        except ValueError:
                            print("Seleccione numeros por favor")
                        else:
                            if LimMax < LimIn or LimMax == LimIn:
                                print("El limite mayor debe ser mas grande que el mayor")
                            elif LimMax > LimIn and LimMax - LimIn == 1:
                                print("La diferencia entre los limites debe ser mayor a 1")
                            else:
                                act8(LimIn,LimMax)
                                print("")
                                break

                case 9:
                    while True:
                        try:
                            ba = int(input("Dame el valor de la base del triangulo: "))
                            al = int(input("Dame el valor de la altura del triangulo: "))
                        except ValueError:
                            print("Escriba unicamente numeros por favor")
                        else:
                            ar = act9(ba,al)
                            break
                    print(f"El area del triangulo son {ar} centimetros cuadrados")
                    print("")

                case 10:
                    act10()
                    print("")

                case 11:
                    break

                case _:
                    print("Error de seleccion\n")


def act1():
    par, inpar = 0,0
    for x in range(40):
        num = random.randint(0,200)
        if num == 0:
            print("0 no es par o impar")
        elif num%2 == 0:
            print(f"{num} es un numero par")
            par = par+1
        else:
            print(f"{num} es un numero impar")
            inpar = inpar + 1
    print(f"\nHubo {par} numeros pares")
    print(f"Hubo {inpar} numeros inpares")

def act2():
    while True:
        try:
            num = int(input("Dame un numero del 1 al 20: "))
        except ValueError:
            print("Seleccione un numero por favor")
        else:
            if num > 0 and num <= 20:
                for x in range(10):
                    print(f"{num} x {x+1} = {num * (x+1)}")
                break
            else:
                print("Numero fuera de rango")

def act3():
    while True:
        try:
            cal = int(input("Dame tu calificacion: "))
        except ValueError:
            print("Hubo un error en la calificacion, intente de nuevo\n")
        else:
            if cal >= 0 and cal <= 59:
                print("Reprobado")
                break
            elif cal >= 60 and cal <= 100:
                print("Aprobado")
                break

            else:
                print("Hubo un error en la calificacion, intente de nuevo\n")


def act4():
    while True:
        try:
            num = int(input("Dame un numero positivo (Escribe 0 para salir): "))
        except ValueError:
            print("Introdusca solamente numeros enteros positivos por favor")
        else:
            if num == 0:
                break
            elif num <= 0:
                print("Favor de introducior solo numeros positivos")
            else:
                print(f"{num} es un numero valido")

def act5():
    opor = 3
    while opor > 0:
        try:
            cal = int(input("Dame tu calificacion: "))
        except ValueError:
            print("Calificacion no valida")
        else:
            if cal > 0 and cal < 60:
                if opor > 1:
                    print("Reprobado")
                    opor = opor - 1

                else:
                    print("Baja academica")
                    opor = opor - 1


            elif cal >= 60 and cal <= 100:
                print("Aprobado")
                break

            else:
                print("Calificaion no valida")

def act6():
    nums = []
    total = 0
    while True:
        try:
            num = int(input("Dame un numero (Seleccionar el numero 0 para salir): "))
        except ValueError:
            print("Elija un numero valido\n")
        else:
            if num == 0:
                break
            elif num > 0:
                nums.append(num)
                total = total + num
            else:
                print("Elija un numero positivo\n")

    maxi = max(nums)
    mini = min(nums)
    media = total/len(nums)
    print(nums)
    print(f"La suma de los numeros es {total}")
    print(f"La media de los numeros es {media}")
    print(f"El numero mas grande fue {maxi}")
    print(f"El numero mas pequenio fue {mini}")

def act7():
    inpar,par,cont = 0,0,0
    nums = []

    while inpar < 15:
            num = random.randint(10,60)
            if num%2 == 0:
                nums.append(num)
                cont = cont + 1
                par = par + 1
            else:
                nums.append(num)
                cont = cont + 1
                inpar = inpar + 1
            if cont > 45:
                break
    print(nums)
    numeros = len(nums)
    print(f"Se generaron {numeros} numeros")
    print(f"hubo {par} numeros pares")
    print(f"hubo {inpar} numeros inpares")


def act8(men,may):
    while True:
        try:
            num = int(input(f"Dame un numero mayor a {men} y menor a {may} (Escriba 724 para salir): "))
        except ValueError:
            print("Este no es un numero")
        else:
            if num > men and num < may:
                print("Este nunero se encuentra dentro del rango")
            elif num == 724:
                break
            else:
                print("Este numero esta fuera de rango")


def act9(bas,alt):
    return(bas * (alt / 2))

def act10():
    lim = 15
    while True:
        try:
            num = int(input("Dame un numero mayor a 20 (0 para salir): "))
        except ValueError:
            print("Este no es un numero")
        else:
            if num > 0 and num < 15:
                print("Este nunero es menor a 15")
            elif num == 0:
                break
            else:
                print("Este numero es valido")

if __name__ == "__main__":
    main()
```
